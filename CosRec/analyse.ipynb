```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
def print_styling(figsize=(10,8)):
    plt.rc('figure', figsize=figsize) 
    plt.style.use("seaborn")
    sns.set_palette("dark")
    SMALL_SIZE = 15
    MEDIUM_SIZE = 18
    BIGGER_SIZE = 26

    plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
    plt.rc('lines', linewidth=2)

    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    
    plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def plot_curve(df,x = "SCALE", y = "s_mean", ax=None, mode = "sklearn", titel = "R2-Score", SaveName = False, y_label = "score"):
    if ax==None:
        ax = plt.gca()
    plt.sca(ax)

    plt.semilogy(df[x], df[y], label = mode, marker='x')

    plt.grid(True)
    plt.legend()
    plt.xlabel("n_iter")
    plt.suptitle(titel)
    if "time" in y:
        plt.ylabel("time [s]")
    else:
        plt.ylabel(y_label)
    if SaveName:
        plt.savefig(SaveName)
```

```python
df_original = pd.read_csv("results/ml1m_original", index_col=None, header=0,sep=",")
df_hold = pd.read_csv("results/ml1m_hold", index_col=None, header=0,sep=",")
#df_CV = pd.read_csv("results/ml1m_CV", index_col=None, header=0,sep=",")
```

```python
mpl.style.use('seaborn')
plt.figure(figsize=(10,8))
print_styling()
plot_curve(df_original,x = "Epoch", y = "loss", mode="ml1m_original",titel = "learning-curve",y_label = "loss")
plot_curve(df_hold,x = "Epoch", y = "loss", mode="ml1m_hold",titel = "learning-curve",y_label = "loss")
plt.savefig("out/Learning_Curve_ml1m")
```
